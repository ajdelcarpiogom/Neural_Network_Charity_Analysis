```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("C:/Users/andre/Documents/Analysis_Projects/Neural_Network_Charity_Analysis/charity_data.csv")
application_df.head()
```

```python
application_df.drop(columns = ['EIN'], inplace = True)
```

```python
counts1 = application_df['NAME'].value_counts()
counts1
```

```python
names_replace = list(counts1[counts1<5].index)
names_replace
```

```python
for name in names_replace:
    application_df['NAME'] = application_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
counts = application_df['APPLICATION_TYPE'].value_counts()
counts
```

```python
application_types_to_replace = list(counts[counts<500].index)
application_types_to_replace
```

```python
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
countsbinning = application_df['CLASSIFICATION'].value_counts()
```

```python
countsclassification = countsbinning[countsbinning>1]
```

```python
classifications_to_replace = list (countsbinning[countsbinning<100].index)
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df = pd.get_dummies(application_df,dtype=float)
```

```python
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop('IS_SUCCESSFUL', axis=1).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len( X_train_scaled[0])
hidden_nodes_layer1=7
hidden_nodes_layer2=14
hidden_nodes_layer3=21
nn = tf.keras.models.Sequential()

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=5)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
